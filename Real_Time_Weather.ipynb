```python
import pandas as pd

# Load datasets
temperature = pd.read_csv("temperature.csv")
humidity = pd.read_csv("humidity.csv")
pressure = pd.read_csv("pressure.csv")
wind_speed = pd.read_csv("wind_speed.csv")
wind_direction = pd.read_csv("wind_direction.csv")
weather_desc = pd.read_csv("weather_description.csv")

# Melt each DataFrame
def melt_df(df, value_name):
    return df.melt(id_vars=["datetime"], var_name="city", value_name=value_name)

temperature = melt_df(temperature, "temperature")
humidity = melt_df(humidity, "humidity")
pressure = melt_df(pressure, "pressure")
wind_speed = melt_df(wind_speed, "wind_speed")
wind_direction = melt_df(wind_direction, "wind_direction")
weather_desc = melt_df(weather_desc, "weather_description")

# Merge all into a single DataFrame
weather_df = temperature \
    .merge(humidity, on=["datetime", "city"], how="outer") \
    .merge(pressure, on=["datetime", "city"], how="outer") \
    .merge(wind_speed, on=["datetime", "city"], how="outer") \
    .merge(wind_direction, on=["datetime", "city"], how="outer") \
    .merge(weather_desc, on=["datetime", "city"], how="outer")

# Convert datetime column to datetime object
weather_df["datetime"] = pd.to_datetime(weather_df["datetime"])

# Preview
weather_df.head()
```

```python
weather_df.isnull().sum()
```

```python
weather_df = weather_df.dropna(thresh=4)  # keep rows with at least 4 non-null values
```

```python
weather_df.fillna(method='ffill', inplace=True)  # forward fill
```

```python
weather_df.to_csv("final_weather_data.csv", index=False)
```
